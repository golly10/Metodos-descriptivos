```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', sep=",")

columns = ["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country", "class"]

data.columns = columns

data = data.drop("fnlwgt", axis=1)
```

```python
data["age"] = pd.cut(data["age"], bins=3, right=False)
data["education-num"] = pd.cut(data["education-num"], bins=3, right=False)
data["capital-gain"] = pd.cut(data["capital-gain"], bins=3, right=False)
data["capital-loss"] = pd.cut(data["capital-loss"], bins=3, right=False)
data["hours-per-week"] = pd.cut(data["hours-per-week"], bins=3, right=False)

data
```

```python
data_encoded = pd.get_dummies(data)

data_encoded
```

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(data_encoded, min_support=0.4, use_colnames=True)
```
